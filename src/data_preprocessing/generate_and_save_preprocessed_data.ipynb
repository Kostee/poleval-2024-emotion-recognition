# Determinte constants

```python
DATA_DIR = 'data/'
IN_PATHS = ['train/in_baseline.tsv', 'train/in_gpt_corr.tsv',
            'test-A/in_baseline.tsv', 'test-A/in_gpt_corr.tsv',
            'test-B/in_baseline.tsv', 'test-B/in_gpt_corr.tsv']
OUT_PATHS = ['train/in_prep_bas.tsv', 'train/in_prep_gpt.tsv',
             'test-A/in_prep_bas.tsv', 'test-A/in_prep_gpt.tsv',
             'test-B/in_prep_bas.tsv', 'test-B/in_prep_gpt.tsv']
```

# Function to preprocess text

```python
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'\W', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

# Load the datasets

```python
datasets = []
for in_path in IN_PATHS:
    datasets.append(pd.read_csv(DATA_DIR + in_path, sep = '\t'))
```

# Apply preprocessing

```python
for i in range(len(datasets)):
    datasets[i]['text'] = datasets[i]['text'].apply(preprocess_text)
```

# Save the cleaned text to new files

```python
for i in range(len(datasets)):
    datasets[i][['text']].to_csv(DATA_DIR + OUT_PATHS[i], sep='\t', index=False)
```

# Display first few rows of each cleaned data

```python
for i in range(len(datasets)):
    print("Top", IN_PATHS[i], "rows:")
    print(datasets[i][['text']].head(), '\n')
```
